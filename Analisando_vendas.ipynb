```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
lojas = ['A', 'B', 'C', 'D']

vendas_2022 = {'Jan': [100, 80, 150, 50],
    'Fev': [120, 90, 170, 60],
    'Mar': [150, 100, 200, 80],
    'Abr': [180, 110, 230, 90],
    'Mai': [220, 190, 350, 200],
    'Jun': [230, 150, 280, 120],
    'Jul': [250, 170, 300, 140],
    'Ago': [260, 180, 310, 150],
    'Set': [240, 160, 290, 130],
    'Out': [220, 140, 270, 110],
    'Nov': [400, 220, 350, 190],
    'Dez': [300, 350, 400, 250]
}
```

```python
dados = pd.DataFrame(vendas_2022, lojas)
dados
```

```python
print(dados.index.name)
```

```python
dados.index.name = 'Lojas'
```

```python
dados
```

```python
dados['Total'] = dados.sum(axis = 1)
dados
```

```python
fig, ax = plt.subplots(figsize = (10,5))
ax.barh(dados.index, dados.Total)
ax.set_title ('Venda do ultimo ano', loc = 'left', fontsize = 15)
ax.set_ylabel('Lojas', fontsize = 12)
ax.set_xlabel('Quantidade de Vendas', fontsize = 12)
ax.xaxis.set_tick_params(labelsize = 11)
ax.yaxis.set_tick_params(labelsize = 11)

for i, v in enumerate(dados.Total):
    ax.text(v + 20, i, str(v), color = 'black', fontsize = 11, ha = 'left', va = 'center')

ax.set_frame_on(False)
ax.get_xaxis().set_visible(False)
ax.tick_params(axis = 'both', which = 'both', length = 0)

plt.show
```

```python
meses = ['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun', 'Jul', 'Ago', 'Set', 'Out', 'Nov', 'Dez']
meses
```

```python
fig, axs = plt.subplots(2,2, figsize = (12,8))
fig.subplots_adjust(hspace = 0.5, wspace = 0.3)
fig.suptitle('Comparativo de vendas do ano de cada loja por mês')

axs[0,0].plot(dados.loc['A', meses])
axs[0,0].set_title('Loja A', fontsize = 15, loc = 'left')

axs[0,1].plot(dados.loc['B', meses])
axs[0,1].set_title('Loja B', fontsize = 15, loc = 'left')

axs[1,0].plot(dados.loc['C', meses])
axs[1,0].set_title('Loja C', fontsize = 15, loc = 'left')

axs[1,1].plot(dados.loc['D', meses])
axs[1,1].set_title('Loja D', fontsize = 15, loc = 'left')

for ax in axs.flat:
    ax.set_xlabel('Mêses', fontsize = 12)
    ax.set_ylabel('Valores vendidos', fontsize = 12)
    ax.xaxis.set_tick_params(labelsize = 11)
    ax.yaxis.set_tick_params(labelsize = 11)   
    ax.grid(linestyle = '--')

ymin = 0
ymax = 500

for ax in axs.ravel():
    ax.set_ylim(ymin, ymax)

plt.show()
```
